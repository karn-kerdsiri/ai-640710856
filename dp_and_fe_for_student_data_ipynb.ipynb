```python
import pandas as pd
```

```python
df = pd.read_csv('student.csv')
```

**ลบแถวที่มีค่าหาย**

```python
df_clean = df.dropna()
```

```python
df_clean
```

**การจัดการค่าผิดปกติ**

```python
Q1 = df['อายุ'].quantile(0.25)
Q3 = df['อายุ'].quantile(0.75)
IQR = Q3 - Q1
```

```python
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
```

```python
df_clean = df_clean[(df_clean['อายุ'] >= lower_bound) & (df_clean['อายุ'] <= upper_bound)]
```

```python
print(f"\nขอบเขตล่าง - {lower_bound}")
print(f"ขอบเขตบน - {upper_bound}")
print(f"\nข้อมูลหลังกรองค่าผิดปกติ")
print(df_clean)
print(f"\nค่าเฉลี่ยอายุ (หลังกรองค่าผิดปกติ) - {df_clean['อายุ'].mean():.2f}")
```

```python
df_clean = df_clean[(df_clean['GPA'] >= 0.00) & (df_clean['GPA'] <= 4.00)]
```

```python
df_clean
```

**การทำให้ข้อมูลเป็นมาตรฐาน**

```python
def standardize_gender(name):
    if name in ['ชาย', 'M', 'ช','Male']:
        return 'ชาย'
    elif name in ['หญิง','ญ','F', 'Female']:
        return 'หญิง'
    else:
        return name
```

```python
df_clean['เพศ'] = df_clean['เพศ'].apply(standardize_gender)
```

```python
df_clean
```

```python
def standardize_faculty(name):
    if name in ['วิทยาศาสตร์', 'วิทย์', 'Sci','คณะวิทยาศาสตร์','Science']:
        return 'วิทยาศาสตร์'
    elif name in ['วิศวกรรมศาสตร์','วิศวฯ','วิศวะ','Engineering']:
        return 'วิศวกรรมศาสตร์'
    elif name in ['ศิลปศาสตร์','Liberal Arts','ศิลป์','Arts']:
        return 'ศิลปศาสตร์'
    else:
        return name
```

```python
df_clean['คณะ'] = df_clean['คณะ'].apply(standardize_faculty)
```

```python
df_clean
```
